# Navigation JS

```python
#| export
from cjm_fasthtml_token_selector.core.config import TokenSelectorConfig
from cjm_fasthtml_token_selector.core.html_ids import TokenSelectorHtmlIds
```

```python
#| export
def _generate_gap_nav_js(
    config:TokenSelectorConfig,  # config for this instance
    ids:TokenSelectorHtmlIds,    # HTML IDs
) -> str:  # JS code fragment
    """Generate gap mode navigation functions."""
    wrap = "true" if config.wrap_navigation else "false"
    return f"""
    ns.moveLeft = function() {{
        if (ns.anchor > 0) {{
            ns.anchor--;
            ns.focus = ns.anchor;
        }} else if ({wrap}) {{
            ns.anchor = ns.wordCount;
            ns.focus = ns.anchor;
        }}
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.moveRight = function() {{
        if (ns.anchor < ns.wordCount) {{
            ns.anchor++;
            ns.focus = ns.anchor;
        }} else if ({wrap}) {{
            ns.anchor = 0;
            ns.focus = ns.anchor;
        }}
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.selectGap = function(position) {{
        ns.anchor = Math.max(0, Math.min(position, ns.wordCount));
        ns.focus = ns.anchor;
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.moveToFirst = function() {{
        ns.anchor = 0;
        ns.focus = ns.anchor;
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.moveToLast = function() {{
        ns.anchor = ns.wordCount;
        ns.focus = ns.anchor;
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};
"""
```

```python
#| export
def _generate_word_nav_js(
    config:TokenSelectorConfig,  # config for this instance
    ids:TokenSelectorHtmlIds,    # HTML IDs
) -> str:  # JS code fragment
    """Generate word mode navigation functions."""
    wrap = "true" if config.wrap_navigation else "false"
    return f"""
    ns.moveLeft = function() {{
        if (ns.anchor > 0) {{
            ns.anchor--;
            ns.focus = ns.anchor;
        }} else if ({wrap} && ns.wordCount > 0) {{
            ns.anchor = ns.wordCount - 1;
            ns.focus = ns.anchor;
        }}
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.moveRight = function() {{
        if (ns.anchor < ns.wordCount - 1) {{
            ns.anchor++;
            ns.focus = ns.anchor;
        }} else if ({wrap}) {{
            ns.anchor = 0;
            ns.focus = ns.anchor;
        }}
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.selectWord = function(index) {{
        ns.anchor = Math.max(0, Math.min(index, ns.wordCount - 1));
        ns.focus = ns.anchor;
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.moveToFirst = function() {{
        ns.anchor = 0;
        ns.focus = ns.anchor;
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.moveToLast = function() {{
        ns.anchor = Math.max(0, ns.wordCount - 1);
        ns.focus = ns.anchor;
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};
"""
```

```python
#| export
def _generate_span_nav_js(
    config:TokenSelectorConfig,  # config for this instance
    ids:TokenSelectorHtmlIds,    # HTML IDs
) -> str:  # JS code fragment
    """Generate span mode navigation functions."""
    wrap = "true" if config.wrap_navigation else "false"
    return f"""
    ns.moveLeft = function() {{
        if (ns.focus > 0) {{
            ns.focus--;
            ns.anchor = ns.focus;
        }} else if ({wrap} && ns.wordCount > 0) {{
            ns.focus = ns.wordCount - 1;
            ns.anchor = ns.focus;
        }}
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.moveRight = function() {{
        if (ns.focus < ns.wordCount - 1) {{
            ns.focus++;
            ns.anchor = ns.focus;
        }} else if ({wrap}) {{
            ns.focus = 0;
            ns.anchor = ns.focus;
        }}
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.extendLeft = function() {{
        if (ns.focus > 0) {{
            ns.focus--;
        }}
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.extendRight = function() {{
        if (ns.focus < ns.wordCount - 1) {{
            ns.focus++;
        }}
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.selectToken = function(index) {{
        ns.focus = Math.max(0, Math.min(index, ns.wordCount - 1));
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.selectGap = function(position) {{
        var idx = Math.max(0, Math.min(position, ns.wordCount - 1));
        ns.anchor = idx;
        ns.focus = idx;
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.moveToFirst = function() {{
        ns.anchor = 0;
        ns.focus = 0;
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};

    ns.moveToLast = function() {{
        var last = Math.max(0, ns.wordCount - 1);
        ns.anchor = last;
        ns.focus = last;
        ns.updateDisplay();
        ns.updateInputs();
        ns._fireOnChange();
    }};
"""
```

## Public API

```python
#| export
def generate_navigation_js(
    config:TokenSelectorConfig,  # config for this instance
    ids:TokenSelectorHtmlIds,    # HTML IDs
) -> str:  # JS code fragment for the IIFE
    """Generate mode-specific navigation and selection JS functions."""
    generators = {
        "gap": _generate_gap_nav_js,
        "word": _generate_word_nav_js,
        "span": _generate_span_nav_js,
    }
    gen = generators.get(config.selection_mode, _generate_gap_nav_js)
    return gen(config, ids)
```

```python
from cjm_fasthtml_token_selector.core.config import TokenSelectorConfig, _reset_prefix_counter
from cjm_fasthtml_token_selector.core.html_ids import TokenSelectorHtmlIds

_reset_prefix_counter()

# Gap mode
cfg = TokenSelectorConfig(prefix="t", selection_mode="gap")
ids = TokenSelectorHtmlIds(prefix="t")
js = generate_navigation_js(cfg, ids)
assert "ns.moveLeft" in js
assert "ns.moveRight" in js
assert "ns.selectGap" in js
assert "ns.moveToFirst" in js
assert "ns.moveToLast" in js
print("Gap mode navigation JS generated!")

# Word mode
cfg_w = TokenSelectorConfig(prefix="tw", selection_mode="word")
js_w = generate_navigation_js(cfg_w, ids)
assert "ns.selectWord" in js_w
print("Word mode navigation JS generated!")

# Span mode
cfg_s = TokenSelectorConfig(prefix="ts", selection_mode="span")
js_s = generate_navigation_js(cfg_s, ids)
assert "ns.extendLeft" in js_s
assert "ns.extendRight" in js_s
assert "ns.selectToken" in js_s
print("Span mode navigation JS generated!")

_reset_prefix_counter()
print("All navigation JS tests passed!")
```
